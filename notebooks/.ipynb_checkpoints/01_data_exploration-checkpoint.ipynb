# HarmonixSet — Exploratory Data Analysis

Exploring the 912 pop tracks with beats, downbeats, and functional segment annotations.

**Dataset**: [Harmonix Set](https://github.com/urinieto/harmonixset) — 912 Western Pop tracks annotated with structural segment functions.

**Purpose**: Understand data characteristics before implementing SpecTNT model.

```python
import sys, os, csv, json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from collections import Counter, defaultdict

sys.path.append(os.path.abspath('..'))
from utils.label_conversion import convert_label, convert_segments, CLASSES, CLASS_TO_IDX

%matplotlib inline
plt.rcParams['figure.dpi'] = 120
plt.rcParams['figure.figsize'] = (14, 5)
```

## 1. Metadata Overview

```python
DATA_DIR = os.path.abspath('../data/harmonixset')
MELSPEC_DIR = os.path.join(DATA_DIR, 'melspecs')
SEGMENT_DIR = os.path.join(DATA_DIR, 'dataset/segments')
AUDIO_DIR = os.path.join(DATA_DIR, 'audio')
METADATA_PATH = os.path.join(DATA_DIR, 'dataset/metadata.csv')
MELSPEC_INFO = os.path.join(MELSPEC_DIR, 'info.json')

with open(MELSPEC_INFO) as f:
    info = json.load(f)
print('Mel-spec parameters:')
for k, v in info.items():
    print(f'  {k}: {v}')
```

## 2. Mel-Spectrogram Inspection

```python
# Load a sample melspec
sample_id = '0001_12step'
mel_path = os.path.join(MELSPEC_DIR, f'{sample_id}-mel.npy')
mel_spec = np.load(mel_path)
print(f'Shape: {mel_spec.shape}')  # (n_mels, n_frames)
print(f'Dtype: {mel_spec.dtype}')
print(f'Range: [{mel_spec.min():.2f}, {mel_spec.max():.2f}]')

# Get duration and fps
sr = info['SR']
hop_length = info['HOP_LENGTH']
n_frames = mel_spec.shape[1]
duration_sec = n_frames * hop_length / sr
fps = sr / hop_length
print(f'Duration: {duration_sec:.1f}s  ({n_frames} frames @ {fps:.1f} fps)')
```

### 2.1 FPS Decision for Model

Paper uses SR=16000, hop=512 → ~31.25 STFT fps, with activation curves at ~5.2 fps.
Our melspecs: SR=22050, hop=1024 → ~21.53 fps.

We'll temporally downsample in the ResNet front-end to ~5.2 fps. Let's verify the ratio.

```python
native_fps = sr / hop_length
target_fps = 5.2
downsample_ratio = native_fps / target_fps
print(f'Native fps: {native_fps:.2f}')
print(f'Target fps: {target_fps:.2f}')
print(f'Downsample ratio needed: {downsample_ratio:.2f}x')
print(f'≈ {downsample_ratio:.0f}x reduction (e.g., 4x stride/pooling)')
```

## 4. Label Conversion (Algorithm 1)

```python
# Apply conversion to all raw labels
converted_counts = Counter()
conversion_map = {}  # raw -> converted
unmapped = set()
for raw_label, count in all_raw_labels.items():
    c = convert_label(raw_label)
    conversion_map[raw_label] = c
    converted_counts[c] += count
    if raw_label.lower() not in ('end',) and c == 'inst' and 'inst' not in raw_label.lower() and 'break' not in raw_label.lower():
        unmapped.add(raw_label)

print('Converted label distribution:')
print(f"{'Class':<15} {'Count':>6} {'%':>8}")
print('-' * 31)
for c in CLASSES + ['end']:
    cnt = converted_counts.get(c, 0)
    print(f'{c:<15} {cnt:>6} {cnt/total*100:>7.2f}%')

print(f'\nUnmapped labels (not "inst" in original, but mapped to inst):')
for u in sorted(unmapped)[:20]:
    print(f'  "{u}" ({all_raw_labels[u]}×)')
```

```python
# Coverage: how many raw labels are captured by the substring rules?
covered = total - converted_counts.get('inst', 0) - all_raw_labels.get('end', 0) - all_raw_labels.get('inst', 0)
inst_from_rules = converted_counts.get('inst', 0) - all_raw_labels.get('inst', 0)
print(f'Total segments: {total}')
print(f'Covered by substring rules (not inst/end): {covered} ({covered/total*100:.1f}%)')
print(f'  - Explicit "inst" labels: {all_raw_labels.get("inst", 0)}')
print(f'  - Mapped to "inst" via rules: {inst_from_rules}')
print(f'  - Mapped to "inst" via fallback: {converted_counts.get("inst", 0) - all_raw_labels.get("inst", 0) - inst_from_rules}')
print(f'"end" labels: {all_raw_labels.get("end", 0)}')
```

## 5. Overlay Segments on Mel-Spectrogram

```python
# Define colors for the 7 classes
CLASS_COLORS = {
    'intro': '#1f77b4',
    'verse': '#ff7f0e',
    'chorus': '#2ca02c',
    'bridge': '#d62728',
    'outro': '#9467bd',
    'inst': '#8c564b',
    'silence': '#7f7f7f',
}

def plot_song_with_segments(song_id, meta, mel_spec, segments_df):
    """Plot mel-spectrogram with segment overlays."""
    n_mels, n_frames = mel_spec.shape
    sr = info['SR']
    hop_length = info['HOP_LENGTH']
    duration = n_frames * hop_length / sr
    
    song_meta = meta[meta['File'] == song_id]
    title = song_meta['Title'].values[0] if len(song_meta) > 0 else song_id
    artist = song_meta['Artist'].values[0] if len(song_meta) > 0 else ''
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(18, 6), gridspec_kw={'height_ratios': [3, 1]})
    
    # Mel spectrogram
    ax1.imshow(mel_spec, aspect='auto', origin='lower',
               extent=[0, duration, 0, n_mels], cmap='magma')
    
    # Overlay segment boundaries
    for _, row in segments_df.iterrows():
        ax1.axvline(x=row['boundary'], color='cyan', linestyle='--', alpha=0.7, linewidth=0.8)
    
    ax1.set_ylabel('Mel band')
    ax1.set_title(f'{title} — {artist} ({song_id})')
    
    # Segment bar below
    for i, (_, row) in enumerate(segments_df.iterrows()):
        label = convert_label(row['label'])
        if label == 'end':
            continue
        start = row['boundary']
        if i + 1 < len(segments_df):
            end = segments_df.iloc[i + 1]['boundary']
        else:
            end = duration
        color = CLASS_COLORS.get(label, '#cccccc')
        ax2.barh(0, width=end-start, left=start, height=1, color=color, edgecolor='white', linewidth=0.5)
        mid = (start + end) / 2
        if end - start > duration * 0.03:
            ax2.text(mid, 0, label, ha='center', va='center', fontsize=9, fontweight='bold', color='white')
    
    ax2.set_xlim(0, duration)
    ax2.set_ylim(-0.5, 0.5)
    ax2.set_xlabel('Time (seconds)')
    ax2.set_yticks([])
    ax2.set_title('Converted Segment Labels')
    
    # Legend
    patches = [mpatches.Patch(color=c, label=l) for l, c in CLASS_COLORS.items()]
    ax2.legend(handles=patches, loc='upper right', ncol=7, fontsize=8, framealpha=0.9)
    
    plt.tight_layout()
    plt.show()
```

```python
# Plot a few songs
for sid in ['0001_12step', '0095_firework', '0168_maneater', '0217_pokerface']:
    mel = np.load(os.path.join(MELSPEC_DIR, f'{sid}-mel.npy'))
    seg = pd.read_csv(os.path.join(SEGMENT_DIR, f'{sid}.txt'), sep=' ', names=['boundary', 'label'])
    plot_song_with_segments(sid, meta, mel, seg)
```

## 6. Structural Statistics

```python
# Aggregate segment-level stats across all songs
all_seg_lengths = defaultdict(list)
all_seg_counts = Counter()
songs_with_class = defaultdict(int)

for sid in seg_ids:
    try:
        df = pd.read_csv(os.path.join(SEGMENT_DIR, f'{sid}.txt'), sep=' ', names=['boundary', 'label'])
        converted = [convert_label(l) for l in df['label']]
        
        class_in_song = set()
        for i in range(len(df) - 1):
            c = converted[i]
            if c == 'end':
                continue
            seg_len = df.iloc[i + 1]['boundary'] - df.iloc[i]['boundary']
            if seg_len > 0:
                all_seg_lengths[c].append(seg_len)
                all_seg_counts[c] += 1
                class_in_song.add(c)
        
        for c in class_in_song:
            songs_with_class[c] += 1
    except:
        pass

print(f"{'Class':<12} {'#Segments':>10} {'#Songs':>8} {'Mean Len':>9} {'Median':>8} {'Std':>8}")
print('-' * 57)
for c in CLASSES:
    lengths = all_seg_lengths.get(c, [])
    if lengths:
        print(f'{c:<12} {all_seg_counts[c]:>10} {songs_with_class[c]:>8} {np.mean(lengths):>8.2f}s {np.median(lengths):>7.2f}s {np.std(lengths):>7.2f}s')
    else:
        print(f'{c:<12} {0:>10} {0:>8} {"-":>9} {"-":>8} {"-":>8}')
```

```python
# Segment length distribution per class
fig, axes = plt.subplots(2, 4, figsize=(16, 6))
axes = axes.flatten()
for i, c in enumerate(CLASSES):
    lengths = all_seg_lengths.get(c, [])
    if lengths:
        axes[i].hist(lengths, bins=30, color=list(CLASS_COLORS.values())[i], edgecolor='white', alpha=0.8)
        axes[i].axvline(np.median(lengths), color='black', ls='--', label=f"median={np.median(lengths):.1f}s")
        axes[i].set_xlabel('Length (s)')
        axes[i].set_ylabel('Count')
        axes[i].set_title(f'{c} (n={len(lengths)})')
        axes[i].legend(fontsize=8)
    else:
        axes[i].text(0.5, 0.5, 'No data', ha='center', va='center', transform=axes[i].transAxes)
        axes[i].set_title(c)
axes[-1].axis('off')
plt.suptitle('Segment Length Distribution by Class', fontsize=14)
plt.tight_layout()
plt.show()
```

## 7. Data Completeness & Quality Checks

```python
# Check for any mismatches between metadata, melspecs, and segments
print('Songs in metadata but missing melspec:')
missing = meta_ids - melspec_ids
print(f'  Count: {len(missing)}')
if missing:
    print(f'  Examples: {sorted(missing)[:5]}')

print()
print('Songs with mismatched segment ↔ melspec duration:')
mismatches = []
for sid in melspec_ids & seg_ids:
    mel = np.load(os.path.join(MELSPEC_DIR, f'{sid}-mel.npy'))
    mel_dur = mel.shape[1] * hop_length / sr
    try:
        df = pd.read_csv(os.path.join(SEGMENT_DIR, f'{sid}.txt'), sep=' ', names=['boundary', 'label'])
        seg_end = df['boundary'].max()
        diff = abs(mel_dur - seg_end)
        if diff > 2.0:
            mismatches.append((sid, mel_dur, seg_end, diff))
    except:
        pass

print(f'  Songs with >2s mismatch: {len(mismatches)}')
if mismatches:
    for sid, md, sd, diff in mismatches[:5]:
        print(f'    {sid}: mel={md:.1f}s seg={sd:.1f}s diff={diff:.1f}s')
```

### 7.1 Audio check — are WAVs alignable to melspecs?

```python
# Check sample audio file if available
if sample_id in audio_ids:
    import librosa
    audio_path = os.path.join(AUDIO_DIR, f'{sample_id}.wav')
    y, sr_audio = librosa.load(audio_path, sr=None, mono=True)
    print(f'Audio: {len(y)} samples @ {sr_audio} Hz = {len(y)/sr_audio:.1f}s')
    print(f'Mel-spec duration: {mel_spec.shape[1] * 1024 / 22050:.1f}s')
    print()
    if abs(len(y)/sr_audio - mel_spec.shape[1] * 1024 / 22050) < 1.0:
        print('✅ Audio and melspec durations are consistent')
    else:
        print('⚠️ Duration mismatch: audio may not align with precomputed melspec')
else:
    print(f'Audio file for {sample_id} not available')
```
